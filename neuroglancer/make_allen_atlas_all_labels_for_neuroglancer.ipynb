# Purpose of this notebook
There are regions in the Allen and Princeton Mouse Brain ontologies that do not have any voxels associated with them in the annotation volume. However, if one wants to use the merge tool to see parent regions, those regions need to be kept in the atlas in case we collapse child regions to them. I had taken out these regions originally but it turns out I should keep them. The search by region tool should be updated to only include regions that have voxels associated with them currently. 


```python
import os,csv
import pandas as pd
import numpy as np
import tifffile

# Point to the csv files for both atlases containing the region names and voxel counts 
csv_file_allen = '/home/ahoag/ngdemo/data/allen_id_table_w_voxel_counts.csv'
```

```python
df_allen = pd.read_csv(csv_file_allen).sort_values("id")
df_allen
```

```python
IDs=df_allen['id'].values
names=df_allen['name'].values
acronyms = df_allen['acronym'].values
```

```python
for ii in range(len(IDs)):
    ID=IDs[ii]
    name=names[ii]
    acronym=acronyms[ii]
    line = f'[{ID}, "{ID}: {acronym}: {name}"],'
    print(line)
```
