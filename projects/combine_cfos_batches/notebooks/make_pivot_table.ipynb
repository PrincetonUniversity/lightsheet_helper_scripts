## make_pivot_table
The purpose of this notebook is to make the pivot table using the new cfos counts dataframe

```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
import edt
import seaborn as sns

%matplotlib inline
```

## Data loading and validation

```python
main_file = '../data/animals_eroded_counts_20201125.csv'
```

```python
# Read in Will's dataframe containing cell counts in each PMA brain region
main_df = pd.read_csv(main_file)
main_df
```

```python
sub_df = main_df[['brain_longname','condition','batch']]
sub_df
```

```python
pivot_df = sub_df.groupby(['condition','batch']).count().pivot_table('brain_longname',index='condition',columns='batch')
pivot_df = pivot_df.fillna(0)
pivot_df = pivot_df.astype('int')
pivot_df
```

```python
savename_pivot = '../data/pivot_20201125.csv'
pivot_df.to_csv(savename_pivot)
```
